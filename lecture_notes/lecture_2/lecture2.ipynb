```python
# Check upper and lower using ISF (prob x <= X)

# ISF = CHECK BASED ON AREA TO THE RIGHT
upper  = sp.stats.norm.isf(0.025,100,5)
lower = sp.stats.norm.isf(0.975,100,5)

# Check results
print(upper)
print(lower)
```

```python
# Left side, lower bound - PPF is for cumulative
lower  = sp.stats.norm.ppf(0.025,100,5)
print(lower)

# Right side - give to the RIGHT 
upper = sp.stats.norm.isf(0.025,100,5)
print(upper)
```

```python
print(f"The 95% confidence interval is between {lower:.2f} and {upper:.2f} for this production process.")
```

```python
# Approach using standard deviation

ex1 = sp.stats.norm.ppf(0.025) # using PPF, you just need
ex1_mean = 100 + 5*(ex1)
print(ex1_mean)

ex2 = sp.stats.norm.isf(0.025, 0, 1) # Area under the curve, mean and standard deviation (1 default, so it's standard normal)
ex2_mean = 100 + 5*(ex2)
print(ex2_mean)
```

```python
# Basket Size:
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

shopping_basket = pd.read_csv("small_dataset.csv")
shopping_basket.head()
```

```python
transaction_list = list(shopping_basket["ItemsPurchased"].apply(lambda t:t.split(",")))
encoder = TransactionEncoder().fit(transaction_list) # Object

encoded_transaction_list = encoder.transform(transaction_list) # apply the object


# 
ideal_format_table = pd.DataFrame(encoded_transaction_list, columns = encoder.columns_)
ideal_format_table.head()


# Format B for each transaction, and depending on the number of products
# We should use Groupby to transaction ID
```

```python
format_b_to_a = shopping_basket.groupby(["Transaction ID"])["ItemsPurchased"].\
apply(lambda t: ','.join(t)).reset_index()

transaction_list = list(format_b_to_a["ItemsPurchased"].apply(lambda t:t.split(",")))
encoder = TransactionEncoder().fit(transaction_list)
encoded_transaction_list = encoder.transform(transaction_list)
ideal_format_table = pd.DataFrame(encoded_transaction_list, columns = encoder.columns_)
ideal_format_table.head()
```

```python
products = ideal_format_table.columns
```

```python
# Check the "True" statement, with highest frequency of True
result_list = []

for i in range(0,len(products)):
    # Add from product list:

    product_list = []
    
    for index, row in ideal_format_table.iterrows():
        # print(f"Index: {index}")
        # print(f"Row values:\n{row}")
        # Check thru each product:
        count = 0
        if row[i] == True:
            
            # Retrieve what other products are True in this row
            basket = row[row == True].index.to_list()
            # Remove the main item from the basket
            basket.remove(products[i])
            # print(basket)
            
            # Append if not empty, to product list
            product_list.append(basket)
        
    # Once loop is done, add to result list...
    result_list.append(product_list)
        # Check Results:
        # print(result_list)
    print(f"For product: {products[i]}: the list is {result_list}")
                
```
